Ecuación de Antoine

```python
import math

# Constantes de Antoine para el agua
A,B,C=8.07131,1730.63,233.426
T=80 #°c
P=10** (A-(B/(C+T)))
print (f"Presión de vapor del agua a {T}°C:{P:.2f} mmHg") #f antes por que se trata de un flotanto y las llaves () son mis variables y  :. (son decimalesque deseamos en nuestra operación )
```

Número de Reynolds

```python
import math

rho=1000 # densidad (Kg/m3)
v=1.2   # velocidad (m/s)
D= 0.05 #diametro (m)
mu= 0.001 # viscosidad (Pa*s)

Re=(rho*v*D)/mu
print (f"Re={Re:.2f}")
```

Uso de condicionales en el Número de Reynolds

```python
if Re<2300:
    print ("flujo laminar")
elif Re>=2300 and Re<4000:
    print ("flujo transitorio")
else:
    print ("flujo turbulento")
```

Bucle while

se usa cuando no sabemos cuantos veces se repétira solo sabemos que queremos repetir mientras se cumpla una condición.

```python
x=0  # número en el cual comienza
while x<5: # while (mientras)
    print ("x vales:",x) # print (imprime el valor)
    x+=1 # se le suma la x para obtener el siguiente valor (rango)
```

**Romper o saltar bucle **


**break** Detiene el bucle antes de que termine

***continue ***salta a la siguiente iteración sin ejecutar lo que sigue en el ciclo

```python
for i in range (1,10):
  if i==5: #Detiene el ciclo al llegar a 5
     break
  print(i)
```

```python
L= [1,2, None,4,None,5]

suma=0

for i in L:
   if i is None: continue #hace que se salte ese dato (None)
   suma+=i
print (suma)
```
