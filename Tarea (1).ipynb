LEY DE LOS GASES

```python
import math
#Añadir valores conocidas
n=2 #mol
T=298 #K
V=2 #L
R=0.082057 # atm*L/ mol K
# Escribir la ecuación
P=(n*R*T)/(V)
print (f"Presión a {T} K:{P:.3f} atm")
```

ECUACIÓN DE ARRHENIUS




```python

A = 5E07 # 1/s
Ea = 75000 # J/mol
R = 8.314  #J/mol·K
T = 350 #K
k=A*math.exp(((-Ea)/(R*T)))
print (f"constante de velocidad:{k:.7f} s-1")
```

ECUACIÓN DE ANTOINE

```python
import math
A,B,C=8.07131,1730.63,233.426
T=80 #°c
P=10** (A-(B/(C+T)))
print (f"Presión de vapor del agua a {T}°C:{P:.2f} mmHg")
```

NÚMERO DE REYNOLDS

```python
import math

rho=1000 # densidad (Kg/m3)
v=1.2   # velocidad (m/s)
D= 0.05 #diametro (m)
mu= 0.001 # viscosidad (Pa*s)

Re=(rho*v*D)/mu
print (f"Re={Re:.2f}")
```

USO DE CONDICIONES: DETERMINACIÓN DEL RÉGIMEN DE FLUJO

```python


Re= float(input("Ingrese el número de Reynolds :"))
if Re< 2300:
     print("El regimen es laminar")
elif Re>=2300 and Re<=4000:
     print ("El regimen es de transición")
else:
    print ("El regimen es turbulento")

```

Uso de condicionales: Estado del agua respecto al punto crítico

```python

Tc= 647.1 #K
Pc= 22.06 #MPa

T= float(input("Ingrese la temperatura en °K:"))
P= float(input("Ingrese la presión en MPa:"))
if T<Tc and P< Pc:
   print("el agua se encuentra en fase líquida o vapor")
elif T>Tc and P>Pc:
   print("el agua está en estado supercrítico")
elif T < Tc and P > Pc:
   print("el agua se encuentra en condiciones de líquido comprimido")
else:
   print("las condiciones son cercanas al punto crítico")
```
