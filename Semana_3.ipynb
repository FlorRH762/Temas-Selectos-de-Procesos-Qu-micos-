Control de flujo: nos ayuda como va a ir resolucionando las operaciones de forma ordenada

if: se ejecuta si la conción es verdadera
elif: se ejecuta si la anterior fue falsa, pero esta verdadera
else se ejecuta si ninguna anterior fue verdadera.No lleva condición

```python
x=-56   #si es verdadero se ejectua y si no la manda nada

if x< 0: # (: condición)
   print('x es negativo')
```
