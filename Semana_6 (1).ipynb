```python
# Importar librerías necesarias
import matplotlib.pyplot as plt

# Crear figura y eje principal
plt.figure(figsize=(16, 8), dpi=150)
# figsize: tamaño de la figura (ancho, alto)
# dpi: resolución en puntos por pulgada

# Línea principal del precio de cierre
plt.plot(df.index, df["Close"], lw=2.5, color='black', label='Precio de cierre')
# df.index -> eje X (fechas)
# df["Close"] -> eje Y (precios)
# lw=2.5 -> grosor de la línea
# color='black' -> color de la línea
# label -> etiqueta para la leyenda

# ==========================
# CÁLCULO DE MÍNIMO Y MÁXIMO
# ==========================

min_val = df["Close"].min()   # Valor mínimo
max_val = df["Close"].max()   # Valor máximo

min_date = df["Close"].idxmin()  # Fecha del valor mínimo
max_date = df["Close"].idxmax()  # Fecha del valor máximo

# ==========================
# MARCAR PUNTOS EN EL GRÁFICO
# ==========================

# Punto mínimo
plt.scatter(min_date, min_val, color='red', s=100, label='Mínimo')
# scatter -> dibuja un punto (x, y)
# color='red' -> color del punto
# s=100 -> tamaño del marcador
# label -> etiqueta del punto

# Punto máximo
plt.scatter(max_date, max_val, color='green', s=100, label='Máximo')

# ==========================
# ANOTACIONES EN LOS PUNTOS
# ==========================

# Anotar el valor mínimo
plt.annotate(f'Mínimo: {min_val:.2f}',
             xy=(min_date, min_val),               # Coordenadas del punto
             xytext=(min_date, min_val * 1.05),    # Posición del texto (ligeramente arriba)
             arrowprops=dict(facecolor='red', shrink=0.05),
             fontsize=12, color='red')

# Anotar el valor máximo
plt.annotate(f'Máximo: {max_val:.2f}',
             xy=(max_date, max_val),
             xytext=(max_date, max_val * 0.95),    # Posición del texto (ligeramente abajo)
             arrowprops=dict(facecolor='green', shrink=0.05),
             fontsize=12, color='green')

# ==========================
# CONFIGURACIÓN DE ESTILO
# ==========================

plt.xlabel('Fecha', fontsize=18, labelpad=10)            # Etiqueta del eje X
plt.ylabel('Precio de Cierre', fontsize=18, labelpad=10) # Etiqueta del eje Y
plt.title('Precio de Cierre - CEMEX', fontsize=22, fontweight='bold', pad=15)  # Título

plt.grid(alpha=0.3)         # Cuadrícula con transparencia
plt.tick_params(labelsize=14)  # Tamaño de etiquetas de los ejes
plt.legend(fontsize=14)     # Mostrar leyenda
plt.tight_layout()          # Ajustar márgenes automáticamente
plt.show()                  # Mostrar el gráfico
```
